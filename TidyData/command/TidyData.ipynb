```python
import pandas as pd
```

## Dane przedstawiają całkowitą liczbę zarażeń na gruźlicę na przestrzeni lat dla odpowiedniej płci i grupy wiekowej


```python
df = pd.read_csv("file.csv")
## Usuwamy kolumny które nie posiadają danych które możemy dobrze przyporządkować
df.drop(columns=['new_sp', 'new_sp_mu', 'new_sp_fu'], inplace=True)
## Zamiana nazw kolumn 
df.rename(inplace=True, columns={'new_sp_m04': 'm04', 'new_sp_m514': 'm514', 'new_sp_m014': 'm014', 'new_sp_m65': 'm65', 
                                 'new_sp_f04': 'f04', 'new_sp_f514': 'f514', 'new_sp_f014': 'f014', 'new_sp_f65': 'f65','new_sp_f1524': 'f1524', 'new_sp_f2534': 'f2534', 
                                 'new_sp_f4554': 'f4554', 'new_sp_f5564': 'f5564', 'new_sp_m1524': 'm1524', 'new_sp_m2534': 'm2534', 'new_sp_m4554': 'm4554', 
                                 'new_sp_m5564': 'm5564', 'new_sp_m3544': 'm3544', 'new_sp_f3544': 'f3544'})
## Scalenie poprzez sex_age, iso2 oraz year stanowia baze
df = pd.melt(df, id_vars=['iso2','year'], value_vars=list(df.columns)[2:],
            var_name="sex_age", value_name='cases')
## Usunięcie lat z nan cases
df.dropna(inplace=True)

## Agregacja tabeli w taki sposób abysmy otrzymali liczbe zarażeń (na przestrzeni lat) dla każdej płci i grupy wiekowej
aggregation_functions = {'cases':'sum'}
df_new = df.groupby(['sex_age'],as_index=False).aggregate(aggregation_functions).reindex(columns=df.columns)
## Usunięcie kraju oraz roku 
df_new.drop(columns=['iso2', 'year'], inplace=True)

## Rozdzielenie pola sex_age na pola sex oraz age oraz konwersja napisu na czytelniejszy
df_new.sex_age = [s[0]+"_"+s[1:] for s in df_new.sex_age]
df_new[['sex', 'age']] = df_new['sex_age'].str.split('_', expand=True)
df_new = df_new[['sex', 'age', 'cases']]

df_new.replace({'04':'0-4','014': '0-14', '1524': '15-24', '2534': '25-34', '3544': '35-44',
            '4554': '45-54', '5564': '55-64', '65': '65+','514':'5-14'}, inplace=True)
df_new['cases'] = df_new.cases.astype(int)


# df_new.to_csv('sex_age_to_cases.csv')  
df_new

```

## DataFrame pokazuje liczbe zachorowań przypadająca dla danego roku


```python
df = pd.read_csv("file.csv") 
df.drop(columns=['new_sp', 'new_sp_mu', 'new_sp_fu'], inplace=True)
## Zamiana nazw kolumn 
df.rename(inplace=True, columns={'new_sp_m04': 'm04', 'new_sp_m514': 'm514', 'new_sp_m014': 'm014', 'new_sp_m65': 'm65', 
                                 'new_sp_f04': 'f04', 'new_sp_f514': 'f514', 'new_sp_f014': 'f014', 'new_sp_f65': 'f65','new_sp_f1524': 'f1524', 'new_sp_f2534': 'f2534', 
                                 'new_sp_f4554': 'f4554', 'new_sp_f5564': 'f5564', 'new_sp_m1524': 'm1524', 'new_sp_m2534': 'm2534', 'new_sp_m4554': 'm4554', 
                                 'new_sp_m5564': 'm5564', 'new_sp_m3544': 'm3544', 'new_sp_f3544': 'f3544'})
df = pd.melt(df, id_vars=['iso2','year'], value_vars=list(df.columns)[2:],
            var_name="sex_age", value_name='cases')
df.dropna(inplace=True)

aggregation_functions = {'cases':'sum'}
df_new = df.groupby(['year'],as_index=False).aggregate(aggregation_functions).reindex(columns=df.columns)
df_new.drop(columns=['iso2', 'sex_age'], inplace=True)

# df_new.to_csv('year_to_cases.csv')  
df_new
```
